```python
import spacy

nlp = spacy.load('en_core_web_sm')
```

### Token Attributes ###
- token.text: Unicode representation of token text
- token.text_with_ws: includes whitespace
- token.idx: the token's character offset (the character position) in doc
- token.sent: the sentence the token belongs to
- token.is_sent_start: returns a Boolean indicating whether the token starts a sentence
- token.lemma_: lemma form of token
- token.ent_type_: entity type
