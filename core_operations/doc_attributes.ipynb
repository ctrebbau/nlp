```python
import spacy
nlp = spacy.load('en_core_web_md')
```

### Doc Container Attributes ###
- doc.ents: a list of span object with the named entities of the text
- doc.sents: an iterator with the sentence segmentation of the text
- doc.text: returns the Unicode representation of the text
- doc.noun_chunks: yields the noun phrases of the text
- doc.lang_: returns the language the doc created
- doc.to_json(): a useful serialization of the Doc object as JSON

```python
doc = nlp('I jump the chord with Ashley in New York.')
```

```python
doc.ents
```

```python
doc = nlp("The sweet brown fox jumped over the fence.")
```

```python
list(doc.noun_chunks)
```

```python
doc.to_json()
```
