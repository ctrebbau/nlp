```python
import spacy
from spacy.symbols import ORTH
```

## Customizing the tokenizer ##

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("lemme that")
print([w.text for w in doc])

special_case = [{ORTH: "lem"}, {ORTH: "me"}]
nlp.tokenizer.add_special_case("lemme", special_case)

print([w.text for w in nlp("lemme that")])
```

```python
print([w.text for w in nlp("lemme!")])
```

### Debugging the tokenizer ###

```python
contraction1 = "Let's go!"
doc = nlp(contraction1)
nlp.tokenizer.explain(contraction1)
```

```python
contraction2 = "Don't!"
doc = nlp(contraction2)
nlp.tokenizer.explain(contraction2)
```

```python
simple_sent = "I eat icecream."
doc = nlp(simple_sent)
nlp.tokenizer.explain(simple_sent)
```
